###  문제 25. 문제 24에서 가져온 데이터의 Q7으로 시작하는 column들을 합쳐보세요.

- Q7이라는 column을 만들고, row별 데이터를 하나의 list로 합쳐보세요.

```python
# Q7이라는 이름의 새로운 열 추가
lang_df['Q7'] = None

# 각 행(row)의 데이터를 하나의 리스트(list)로 합치는 함수 정의
def row_to_list(row):
    row_list = row.tolist()
    return row_list

# apply 함수를 이용하여 각 행(row)에 대해 row_to_list 함수 적용
lang_df['Q7'] = lang_df.apply(row_to_list, axis=1)

print(lang_df)
```

### 문제 30. 문제 29에서 생성한 데이터를 빈도순으로 정렬하고 color palette를 Set2로 지정해서 출력해보세요.

```python
def has_non_nan(row, cols):
    for col in cols:
        lst = row[col]
        if any(x is not np.nan for x in lst):
            return True
    return False

cols = ['Q17','Q18','Q19','Q20']
mask = method_df.apply(has_non_nan, cols=cols, axis=1)
method_res_df = method_df[mask]
```
